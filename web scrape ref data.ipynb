```python
from bs4 import BeautifulSoup
from pymongo import MongoClient
import pandas as pd
import numpy as np
import requests
import pickle
```

```python
client = MongoClient()
```

```python
client.list_database_names()
```

```python
boxscoresunified = client.boxscoresunified
boxscores = client.boxscores
spreadscombined = client.spreadscombined
finalboxscores = client.finalboxscores
```

```python
teams    =    ['ATL','BOS','BRK','CHI','CHO','CLE','DAL','DEN','DET','GSW','HOU','IND','LAC','LAL','MEM','MIA','MIL','MIN','NOP','NYK',\
              'OKC','ORL', 'PHI','PHO','POR','SAC','SAS','TOR','UTA','WAS']
```

```python
df = pd.DataFrame(list(spreadscombined.unified.find()))
```

```python
soup = BeautifulSoup(df.content[0],'html.parser')
```

```python
#create webscrapper for nba ref stats
```

```python
for team in teams:
    for row in boxscores[team].find():
        row['team'] = team
        finalboxscores.scores.insert_one(row)
```

```python
df = pd.DataFrame(list(finalboxscores.scores.find()))
```

```python
#there is some discrepency amount teams because of the 2012 season where the web scrapper took boxscore links for the first 82 instances
#normally there are 82 regular season games however in 2012 there was a shortened season and only 66 games were played
#So some teams have between 1-16 extra games depending on how many playoff games they played
```

```python
df.year == '2018'
```

```python
final_df = pd.DataFrame()
years = ['2014','2015','2016','2017','2018']
```

```python
for team in teams:
    for year in years:
        final_df = pd.concat([final_df,df[(df.team==team) & (df.year == year)]],axis=0)
final_df
```

```python
final_df.sort_values(['team','year'],inplace=True)
```

```python
final_df.index = range(len(final_df))
```

```python
final_df
```

```python
final_df.to_pickle('final_boxscore_df_2014_2018.pkl')
```
